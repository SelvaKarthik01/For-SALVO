```python
data = np.array(data)
m,n = data.shape
np.random.shuffle(data)

data_dev = data[0:1000].T
y_dev = data_dev[0]
x_dev = data_dev[1:n]
x_dev = x_dev/255.

data_train = data[1000:m].T
y_train = data_train[0]
x_train = data_train[1:n]
x_train = x_train/255.
print(x_train.shape)
```

```python
def initialize():
    w1 = np.random.rand(10,784) - 0.5
    b1 = np.random.rand(10,1) - 0.5
    w2 = np.random.rand(10,10) - 0.5
    b2 = np.random.rand(10,1) - 0.5
    return w1,b1,w2,b2
```

```python
def ReLU(z):
    return np.maximum(z,0)
```

```python
def softmax(z):
    return np.exp(z)/ sum(np.exp(z))
```

```python
def forward_propogation(w1,b1,w2,b2,x):
    z1 = w1.dot(x) + b1
    a1 = ReLU(z1)
    z2 = w2.dot(a1)+ b2
    a2 = softmax(z2)
    return z1,a1,z2,a2
```

```python
def deriv_ReLU(z):
    return z > 0;
```

```python
def back_propogation(z1,a1,z2,a2,w1,w2,x,y):
    linear_y = np.zeros((y.size,y.max()+1))
    linear_y[np.arange(y.size),y] = 1
    linear_y = linear_y.T
    dz2 = a2 - linear_y
    dw2 = (1/m)*dz2.dot(a1.T)
    db2 = (1/m) *np.sum(dz2)
    dz1 = w2.T.dot(dz2) * deriv_ReLU(z1)
    dw1 = 1/m * dz1.dot(x.T)
    db1 = 1/m * np.sum(dz1)
    return dw1,db1,dw2,db2
```

```python
def update_params(w1,b1,w2,b2,dw1,db1,dw2,db2,alpha):
    w1 = w1 - alpha*dw1
    b1 = b1 - alpha*db1
    w2 = w2 - alpha*dw2
    b2 = b2 - alpha*db2
    return w1,b1,w2,b2
```

```python
def get_predictions(a2):
    return np.argmax(a2,0)
def get_accuracy(predictions,y):
    print(predictions,y)
    return np.sum(predictions == y) / y.size
```

```python
def gradient_descent(x,y,alpha,iterations):
    w1,b1,w2,b2 = initialize()
    for i in range(iterations):
        z1,a1,z2,a2 = forward_propogation(w1,b1,w2,b2,x)
        dw1,db1,dw2,db2 = back_propogation(z1,a1,z2,a2,w1,w2,x,y)
        w1,b1,w2,b2 = update_params(w1,b1,w2,b2,dw1,db1,dw2,db2,alpha)
        if i% 10 == 0:
            print("Iteration : ",i)
            print("Accuracy : ",get_accuracy(get_predictions(a2),y))
    return w1,b1,w2,b2
    
    
```

```python
w1,b1,w2,b2 = gradient_descent(x_train,y_train,0.10,1000)
```
